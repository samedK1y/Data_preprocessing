```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns 
```

```python
df = pd.read_csv('heart.csv')

df.head()
```

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
Q1_chl = df['chol'].quantile(0.25)
Q3_chl = df['chol'].quantile(0.75)
IQR_chl = Q3_chl - Q1_chl

alt_sinir_chl = Q1_chl - 1.5*IQR_chl
ust_sinir_chl = Q3_chl + 1.5*IQR_chl

print(f"Kolestrol icin ust sinir: {ust_sinir_chl}")
print(f"Bu değerden büyük olanlar aykırı: {df[df['chol'] > ust_sinir_chl]['chol'].values}")

plt.figure(figsize = (3,3))
sns.boxplot(x = df['chol'], color= 'skyblue')
plt.title("Kolestrolun aykırı degeri")
plt.show()
```

```python

Q1_bst = df['trestbps'].quantile(0.25)
Q3_bst = df['trestbps'].quantile(0.75)
IQR_bst = Q3_bst - Q1_bst

alt_sinir_bst = Q1_bst - 1.5 * IQR_bst
ust_sinir_bst = Q3_bst + 1.5 * IQR_bst  

print(f"Tresebst için üst sınır değer: {ust_sinir_bst}")
print(f"Tresebst için aykırı değerler: {df[df['trestbps'] > ust_sinir_bst]['trestbps'].values}")

plt.figure(figsize= (3,3))
sns.boxplot(x = df['trestbps'], color="skyblue")
plt.title("Trestbps için aykırılık kontrolü")
plt.show()
```

```python
Q1_tha = df['thalach'].quantile(0.25)
Q3_tha = df['thalach'].quantile(0.75)
IQR_tha = Q3_tha - Q1_tha

alt_sinir_tha = Q1_tha - 1.5 *IQR_tha
ust_sinir_tha = Q3_tha + 1.5 *IQR_tha

print(f"Thalach için üst sınır değer: {ust_sinir_tha}")
print(f"Thalach için aykırı değerler: {df[df['thalach'] > ust_sinir_tha]['thalach'].values}")
```

```python
column_chl = 'chol'
df[column_chl] = df[column_chl].clip(alt_sinir_chl,ust_sinir_chl)
print(f"{column_chl} aykırı değerleri baskılandı.")

plt.figure(figsize = (2,2))
sns.boxplot(x = df['chol'], color= 'skyblue')
plt.title("Chol yeni deger grafiği")
plt.show()
```

```python
column_bst = 'trestbps'
df[column_bst] = df[column_bst].clip(alt_sinir_bst,ust_sinir_bst)

print(f"{column_bst} aykırı degerleri baskilandi")

plt.figure(figsize = (3,3))
sns.boxplot(x = df['trestbps'], color= 'skyblue')
plt.title("Tresbst yeni deger grafiği")
plt.show()
```
